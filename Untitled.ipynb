```python
def generate_new_sheet():

    rownames = ["Only 1s",
                "Only 2s",
                "Only 3s",
                "Only 4s",
                "Only 5s",
                "Only 6s",
                "Bonus",
                "Triples",
                "Quadruples",
                "Small street",
                "Big street",
                "Full house",
                "Kniffel",
                "Chance",
                "Upper Sum",
                "Lower Sum",
                "Total"
                ]
    dummy_seq = [0]*len(rownames)
    sheet = pd.DataFrame({"Game 1": dummy_seq, "Game 2": dummy_seq, "Game 3": dummy_seq, "Game 4": dummy_seq, "Game 5": dummy_seq, "Game 6":  dummy_seq})
    sheet.index = rownames
    
    return sheet
```

```python
def generate_empty_eval():
    
    rownames = ["Only 1s",
            "Only 2s",
            "Only 3s",
            "Only 4s",
            "Only 5s",
            "Only 6s",
            "Bonus",
            "Triples",
            "Quadruples",
            "Small street",
            "Big street",
            "Full house",
            "Kniffel",
            "Chance",
            "Upper Sum",
            "Lower Sum",
            "Total"
            ]
    dummy_seq = [0]*len(rownames)
    eval_df = pd.DataFrame({"Hand": dummy_seq, "Score": dummy_seq, "Vector": dummy_seq, "Weighted_Score": dummy_seq})
    eval_df.index = rownames
    
    return eval_df

def first_roll():
    
    dices = list()

    while len(dices) < 5:

        dices.append(random.randint(1, 6))
        
    return dices

def reroll(dice_seq, reroll_pos):
    
    for i in range(len(dice_seq)):
        
        if not reroll_pos[i]:
            
            dice_seq[i] = random.randint(1, 6)
            
    return dice_seq

def eval_hand(dices, eval_df):
    
    return(eval_df)


def check_hand(dice_seq, eval_df):
    
    global rownames
    
    for i in range(6):
    
        vec = list()
        
        for val in dice_seq:
            
            if val == i+1:
                
                vec.append(i+1)
            
            else:
                
                vec.append(0)
    
        # Evaluate upper half
        
        count = dice_seq.count(i+1)
        score = count*(i+1)
        eval_df.iloc[i] = [score, score/(5*(i+1)) + (i+1)/100, str(vec), score * (score/(5*(i+1)))] 
    
        # Check triplets
    
        if count >= 3:
    
            vec = list()

            for val in dice_seq:

                if val == i+1:

                    vec.append(i+1)

                else:

                    vec.append(0)
            
            eval_df.iloc[7] = [np.sum(dice_seq), score/(5*(i+1)), str(vec), score * score/(5*(i+1))] 
            
#         else:
            
#             eval_df.iloc[7] = [0, 0, 0]
    
        # Check quadruplets
    
        if count == 4:
    
            vec = list()

            for val in dice_seq:

                if val == i+1:

                    vec.append(i+1)

                else:

                    vec.append(0)
            
            eval_df.iloc[8] = [np.sum(dice_seq), (score/(5*(i+1))), str(vec), score * score/(5*(i+1))] 
            

    if dice_seq[0] == dice_seq[1] == dice_seq[2] == dice_seq[3] == dice_seq[4]:
        
        eval_df.iloc[12] = [50, score/(5*(i+1)) + 100, str(dice_seq), 100] 
       # print(eval_df.iloc[12])
    
    return eval_df

def correct_scores(eval_df):
    
    for idx in eval_df.index.tolist():
        
        if "Only" in idx:
        
            if eval_df.loc[idx, "Score"] < 0.4:
            
                eval_df.loc[idx, "Score"] = eval_df.loc[idx, "Score"]/10
        
        if "es" in idx:
            
            eval_df.loc[idx, "Score"] = eval_df.loc[idx, "Score"] * (eval_df.loc[idx, "Hand"]/30)
            
            if idx == "Triples" and eval_df.loc[idx, "Hand"] <= 15:
                
                eval_df.loc[idx, "Score"] = eval_df.loc[idx, "Score"]/10
            
            if idx == "Quadruples" and eval_df.loc[idx, "Hand"] <= 20:
                
                eval_df.loc[idx, "Score"] = eval_df.loc[idx, "Score"]/10
            
#     print(eval_df)        
    return(eval_df)
            
def update_boni(df):
    
    for col in df.columns:
        
        if sum(df[col].values[0:6]) > 63:
            
            df[col].values[6] = 35
        
        df[col].values[14] = sum(df[col].values[0:7])
        df[col].values[15] = sum(df[col].values[7:14])
        df[col].values[16] = df[col].values[14] + df[col].values[15]   
        
    return(df)

update_boni(sheet)
```
